```python
import torch
import torch.nn as nn
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()

        self.block1 = nn.Sequential(
            # input channels= 3(RGB), output channels=32(no of filters), kernel size=3(3x3),
            # stride=1(move filter by 1), padding=1(add only one '0' padding)
            nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.block2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.block3 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )

        self.classifier = nn.Sequential(
            nn.Flatten(),
            # 128 channels, 28x28 feature map size after 3 max pooling layers (each reduces the size by half)
            # why 128? because we are using 1 hidden layer with 256 neurons,
            # if it has 1 hidden layer, the output of the hidden layer will be 256,
            # and the output of the output layer will be 10
            nn.Linear(128 * 28 * 28, 256), 
            nn.ReLU(),
            nn.Linear(256, 10)  # 10 output classes for classification
        )

    def forward(self, x):
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        x = self.classifier(x)
        return x # output of the model is a [1,10] tensor,
        # which is the output of the output layer, 
        # and it contains the probabilities of each class
        # like [0.1, 0.2, 0.3, 0.4, 0, 0.6, 0.8, 0, 0, 0.5] for a 10 class classification problem
```
